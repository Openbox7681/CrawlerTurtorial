```python
# 建立 form data
form_data = {"startStation":"977abb69-413a-4ccf-a109-0272c24fd490",
             "endStation":"fbd828d8-b1da-4b06-a3bd-680cdca4d2cd",
             "theDay":"2019/09/13",
             "timeSelect":"17:00",
             "waySelect":"DepartureInMandarin"}


# 將 form_data 透過 post 的方式進行 requests
response_post = requests.post("https://m.thsrc.com.tw/tw/TimeTable/SearchResult"\
                , data = form_data)
soup_post = BeautifulSoup(response_post.text, "lxml")

soup_post = soup_post.find_all("div" , {"class":"timeResultList ui-grid-b"} )[0]

for item in soup_post.find_all("div" , {"class" : "ui-block-b"}):
    input_string = item.getText().replace(" ","").strip()
    if u"出發" not in input_string:
        print(input_string)
```
